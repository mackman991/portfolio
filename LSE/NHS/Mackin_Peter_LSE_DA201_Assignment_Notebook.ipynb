### Breakdown of appointments by duration

```python
# Group and sum duration of appointments
ad_duration_grouped = (
    ad[['actual_duration', 'count_of_appointments']]
    .groupby('actual_duration')
    .sum()
    .reset_index()
)

# Set order
duration_order = [
    '1-5 Minutes',
    '6-10 Minutes',
    '11-15 Minutes',
    '16-20 Minutes',
    '21-30 Minutes',
    '31-60 Minutes',
    'Unknown / Data Quality'
]

# Set category order
ad_duration_grouped['actual_duration'] = pd.Categorical(
    ad_duration_grouped['actual_duration'],
    categories=duration_order,
    ordered=True
)

# Sort by order
ad_duration_grouped = ad_duration_grouped.sort_values('actual_duration')

# View result
print(ad_duration_grouped)

# Plot chart
plt.figure(figsize=(15, 12))
sns.barplot(
    data=ad_duration_grouped,
    x='actual_duration',
    y='count_of_appointments',
    color='#0072CE'
)
plt.title('Total Appointments by Actual Duration')
plt.xlabel('Actual Duration')
plt.ylabel('Count of Appointments')
plt.xticks(rotation=45)
plt.tight_layout()

# Save chart
plt.savefig('appointment_by_actual_duration.jpg')
# Show chart
plt.show()
```

### Frequency distribution of total daily appointments

We decided to remove saturdays and sundays from our analysis, because it skewed the results. Additionally, we removed outliers on weekdays when they fell outside of the lower/upper quartile range.

```python
daily = (
    nc
    .groupby('appointment_date', as_index=False)['count_of_appointments']
    .sum()
    .rename(columns={'count_of_appointments': 'daily_total'})
)

# Remove Saturdays & Sundays
daily['day_of_week'] = daily['appointment_date'].dt.day_name()
daily = daily[~daily['day_of_week'].isin(['Saturday','Sunday'])]

# Calculate z-scores
mean = daily['daily_total'].mean()
std  = daily['daily_total'].std()
daily['z'] = (daily['daily_total'] - mean) / std

# Remove by IQR 
Q1  = daily['daily_total'].quantile(0.25)
Q3  = daily['daily_total'].quantile(0.75)
IQR = Q3 - Q1
lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
clean_iqr = daily[
    (daily['daily_total'] >= lower) &
    (daily['daily_total'] <= upper)
].copy()

# Compare raw vs cleaned distributions
for name, df_clean in [('Raw', daily), ('IQR-filtered', clean_iqr)]:
    plt.figure(figsize=(8,4))
    sns.histplot(df_clean['daily_total'], bins=30, kde=True, edgecolor='w', color='#0072CE')
    plt.title(f'{name} Distribution of Daily Totals')
    plt.xlabel('Daily Appointments')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.savefig('nc_distribution_of_daily_totals{name}.jpg', dpi=300)
    plt.show()

daily_final = clean_iqr
```

### Scenario analysis

We adjusted daily appointment volumes by missed appointments to gain a better understanding of attended appointments.

```python
# Count of weekday-days per month
daily['month'] = daily['appointment_date'].dt.to_period('M')
daily['days_in_month'] = daily.groupby('month')['appointment_date'].transform('count')

daily = daily[daily['daily_total'] >= 200_000].copy()

# Calculate each months total
daily['monthly_total'] = daily.groupby('month')['daily_total'].transform('sum')

# Subtract no-shows for both rates
for rate in (0.03, 0.06):
    col_ns = f'no_shows_{int(rate*100)}p'
    col_adj = f'daily_adj_{int(rate*100)}p'
    daily[col_ns]  = daily['monthly_total'] * rate / daily['days_in_month']
    daily[col_adj] = (daily['daily_total'] - daily[col_ns]).round().astype(int)


# Plot three distributions
plt.figure(figsize=(15,12))
sns.histplot(daily['daily_total'],
             bins=20, kde=True, edgecolor='w', label='Original', color='#768692')
sns.histplot(daily['daily_adj_3p'],
             bins=20, kde=True, edgecolor='w', alpha=0.5, label='Adjusted (3% DNA)', color='#003087')
sns.histplot(daily['daily_adj_6p'],
             bins=20, kde=True, edgecolor='w', alpha=0.5, label='Adjusted (6% DNA)', color='#0072CE')
plt.title('Weekday Daily Appointments: Original vs Adjusted (3% & 6% DNA)')
plt.xlabel('Appointments per Day')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.savefig('nc_adjusted_distribution.jpg', dpi=300)
plt.show()
```
